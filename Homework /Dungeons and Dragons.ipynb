```python
import random

import matplotlib.pyplot as plt


def roll(num_sides, num_rolls):
    dict = {}
    for i in range(1, num_sides+1):
        dict[i] = 0
    for i in range(0, num_rolls):
        num = random.randint(1,num_sides)
        dict[num] += 1
    print(dict)
    
    x_values = list(dict.keys())
    y_values = list(dict.values())
    plt.bar(x_values, y_values, width=1, align='center', color='blue', edgecolor='black')
    plt.xticks(x_values)
    plt.xlabel('Sides')
    plt.ylabel('Roll Count')
    plt.title('Histogram of role count for sides of die')
    plt.show()

roll(20, 1000000)
```

```python
def max_num(n,m):
    l = []
    for i in range(0,m):
        num = random.randint(1,n)
        l.append(num)
        
    return max(l)
```

```python
i = 0
d = {}
dict = {}
for i in range(1, 21):
    dict[i] = 0
        
for i in range(0,1000000):
    num = max_num(20,2)
    dict[num] += 1

x_values = list(dict.keys())
y_values = list(dict.values())
plt.bar(x_values, y_values, width=1, align='center', color='blue', edgecolor='black')
plt.xticks(x_values)
plt.xlabel('Sides')
plt.ylabel('Roll Count')
plt.title('Histogram of role count for sides of die')
plt.show()
```
