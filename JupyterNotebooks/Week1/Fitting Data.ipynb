```python
#Worked with Maddie Holt

import numpy as np
import matplotlib.pyplot as plt
```

```python
def fitting_function(t, a0, a1, a2, a3, a4):
    return a0 + a1 * np.array(t) + a2 * np.sin(a4 * np.array(t)) + a3 * np.cos(a4 * np.array(t))

def fitting_function2(t, a0, a1, a2, a3, a4):
    return a0 + a1 * np.array(t) + a2 * np.sin(a3 * np.array(t) + a4)
```

```python
def gen_data(a0, a1, a2, a3, a4, size, sigma, low, high):
    x = np.random.uniform(low, high, size)
    y = fitting_function(x, a0, a1, a2, a3, a4) + np.random.normal(0, sigma, size)
    return x, y
```

```python
a0 = 0
a1 = 1
a2 = 1
a3 = 1
a4 = 0

sigma = 0.5
size = 100
t_low = 20
t_high = 35

x, y = gen_data(a0, a1, a2, a3, a4, size, sigma, t_low, t_high)

plt.plot(x,y, 'o')
```

```python
from scipy.optimize import curve_fit
popt1, pcov1 = curve_fit(fitting_function, x, y)
popt2, pcov2 = curve_fit(fitting_function2, x, y)

print(popt1)
print(popt2)
```

```python
plt.plot(x,y, 'o')
plt.plot(x, fitting_function(x, *popt1), 'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['data', 'fitting function 1'])
```

```python
plt.plot(x,y, 'o')
plt.plot(x, fitting_function2(x, *popt2), 'g-')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['data', 'fitting function 2'])
```
