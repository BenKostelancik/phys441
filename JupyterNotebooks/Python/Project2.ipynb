```python
def federal_tax(P1,P2): 
    if P2 == 'S':
        if P1 <= 10000.00: 
            result = P1*.05
        elif P1 <= 50000.00:
            result = 500 + (P1 - 10000)*.105
        else:
            result = 4700 + (P1 - 50000)*.2025
    else:
        if P1 <= 25000.00:
            result = P1*.0625 
        elif P1 <= 90000.00:
            result = 1562.50 + (P1 - 25000)*.1075 
        else:
            result = 8550 + (P1 - 90000)*.225 
            
    return result
```

```python
def format_name(name):

    name_split = name.split(" ")

    good_input_format = True

    if (len(name_split) == 3 or len(name_split) == 2):
        for substring in name_split:
            if substring.isalpha():
                continue
            else:
                good_input_format = False
    else:
        good_input_format = False
    
    if good_input_format and len(name_split) == 2:
        return_string = ('%s, %s' %(name_split[1],name_split[0]))
    else:
        if good_input_format and len(name_split) == 3:
            return_string = ('%s, %s %s.' %(name_split[2],name_split[0],name_split[1][0]))
        else:
            name_string = ('The input does not have the correct form')
        
    return return_string
```

```python
def healthcare(P1):
    if P1 == 'S':
        result = 800.0/52.0
    else:
        result = 1000.0/52.0
            
    return result
```
