```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def fx(x):
    return x*np.sin(x)
```

```python
n = 60
x = np.linspace(0,4.0*np.pi,n+1)

y = np.zeros(n+1)

for i in range(n+1):
    y[i] = fx(x[i])
    
print(x,y)
```

```python
def P(x,y,n,xval):

    sum = 0.0
    for j in range(n+1):
        numerator_product = 1.0
        denominator_product = 1.0
        for k in range(n+1):
            if not k==j:
                numerator_product=numerator_product*(xval-x[k])
                denominator_product=denominator_product*(x[j]-x[k])
        sum = sum + numerator_product*y[j]/denominator_product
    
    if abs(sum)>100:
        print ('Weird Point',xval,sum)
        sum = 0.0
    return sum
            
```

```python
xfit = np.linspace(0,4.0*np.pi,n)
yfitP = np.zeros(n)
yfitE = np.zeros(n)
for i in range(n):
    yfitP[i]=P(x,y,n,xfit[i])
    yfitE[i]=fx(xfit[i])
    
#print (yfitP)
#print (yfitE)
```

```python
fig = plt.figure(figsize=(7, 7))
plt.scatter(xfit,yfitE)
plt.plot(xfit,yfitP,'r-')
```
