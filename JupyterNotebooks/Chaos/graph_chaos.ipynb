```python
from array import array
import matplotlib.pyplot as plt
```

```python
def growthfunction(x,par):
    return par[0]*x*(1.0-x)
```

```python
lambda_start =3.5
npoints = 80
xstart = 0.5
```

```python
population = array('d')
year = array('d')
lmbda = array('d')
xvalue = array('d')

population.append(xstart)
lmbda.append(lambda_start)
year.append(0)

n = npoints
for i in range(1,n):
    year.append(i)
    x = population[i-1]
    population.append(growthfunction(x,lmbda))
```

```python
# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 8))

# Set the grid lines
ax.grid(True)

# Set the background color
ax.set_facecolor('#F0F0F0')

# Set the border properties
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set the canvas size
ax.set_title('Simple Chaos Example')

# Set the line and marker properties
line_color = 'r'
line_width = 2
marker_color = 'b'
marker_style = 'o'
marker_size = 5

# Create the plot
plt.plot(year, population, color=line_color, linewidth=line_width,
         marker=marker_style, markersize=marker_size, markerfacecolor=marker_color)

# Set the plot title and axis labels
plt.title("Growth Function: lambda*x*(1-x)")
plt.xlabel("Year")
plt.ylabel("Population")
```
