```python
import numpy as np
import matplotlib.pyplot as plt

nterms = 10000001

nplot = []
splot = []
aplot = []

inf_sum = 0.0
for n in range(nterms):
    inf_sum += np.abs(np.sin(n*1.0))**n
    if n%100 == 0:
        nplot.append(n)
        splot.append(inf_sum)
        aplot.append(np.pi/2.0*np.sqrt(n))

print(f"Final value: {inf_sum}")

plt.plot(nplot,splot)
plt.plot(nplot,aplot)
```
