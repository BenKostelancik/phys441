#### Simple Chaos Example

Population growth is a classic example of a chaotic system.  The population of a species is a function of the population in the previous year.  The function is non-linear and can be written as:

$$x_{n+1} = \lambda x_n (1-x_n)$$

where $x_n$ is the population in year $n$ and $\lambda$ is a constant that depends on the species and the environment.  The function is non-linear because of the quadratic term $x_n^2$.

The first term, $\lambda x_n$, is the growth term.  It is the growth rate times the population.  The second term, $\lambda x_n^2$, is the death term.  It is the death rate times the population squared.  The death term is quadratic because the death rate is assumed to be proportional to the population.

We begin by defining a function that computes the population in year $n+1$ given the population in year $n$ and the value of $\lambda$.

```

````python
def growthfunction(x,par):
    return par[0]*x*(1.0-x)
````

Define the initial population and the value of $\lambda$.  Note that $x_0$ is the initial population expressed as a fracton of the maximum population that possible, i.e.

$$0 < x_n < 1$$

Based on this assumption, it is easy to see that the allowed values of $\lambda$ are:

$$ 0 < \lambda < 4$$

````python
def simulate(lambda_start, npoints, xstart):
    population = array('d')
    year = array('d')
    lmbda = array('d')
    xvalue = array('d')

    population.append(xstart)
    lmbda.append(lambda_start)
    year.append(0)

    n = npoints
    for i in range(1,n):
        year.append(i)
        x = population[i-1]
        population.append(growthfunction(x,lmbda))
        
    # Create a figure and axes
    fig, ax = plt.subplots(figsize=(10, 8))

    # Set the grid lines
    ax.grid(True)

    # Set the background color
    ax.set_facecolor('#F0F0F0')

    # Set the border properties
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Set the canvas size
    ax.set_title('Simple Chaos Example')

    # Set the line and marker properties
    line_color = 'r'
    line_width = 2
    marker_color = 'b'
    marker_style = 'o'
    marker_size = 5

    # Create the plot
    plt.plot(year, population, color=line_color, linewidth=line_width,
         marker=marker_style, markersize=marker_size, markerfacecolor=marker_color)

    # Set the plot title and axis labels
    plt.title("Growth Function: lambda*x*(1-x)")
    plt.xlabel("Year")
    plt.ylabel("Population")

    print(population[0:8])
    
    plt.show()
````

````python
lambda_start = 0.8
npoints = 20
xstart = 0.5

simulate(lambda_start, npoints, xstart)
````

We see that the population oscillates between four distinct values: 0.5, 0.875, 0.3828125, and 0.826934814453125.  The population oscillates between these values because the function is non-linear.  The oscillations are periodic.  The period is four years.  The population in year $n$ is the same as the population in year $n+4$.
