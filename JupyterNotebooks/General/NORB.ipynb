# Linear Regression
using data to see if rating of review and number of images and price effect sales.

```python
#Correlation of the variables
```

```python
x = df_norb[['Number_images']]
y = df_norb[['MonthlySales']]

regressor = LinearRegression()
regressor.fit(x, y)
y_pred = regressor.predict(x)
y_pred
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.title('Number of images vs Monthly Sales')
plt.xlabel('Number of Images')
plt.ylabel('Monthly Sales')
plt.show()
```

There are two problems with the plot above:

1.  The two data points (row indices 28 and 29) with the very large monthly sales appears to be outliers.  So, we want to remove them from the data set, and then refit the data.

2.  The linear fit that we have chosen is of the form: y = mx + b.  That is, it allows for a negative y-intercept.  However, in this case, we cannot have monthly sales less than zero!  So, what we need to do is choose a different type of linear fit - one which fixes the y-intercept to be zero, and does a linear fit of the form: y = mx.

```python
# First, remove the outliers, in reverse order so that we don't have reindexing problems.

df_norb = df_norb.drop([29])
df_norb = df_norb.drop([28])
df_norb
```

```python
# Now, redo the fit.  We specify an additional directive:  fit_intercept=False to remove the fitting of the
# y-intercept
#
# We see that the linear regression fit looks more reasonable now! :)

x = df_norb[['Number_images']]
y = df_norb[['MonthlySales']]

regressor = LinearRegression(fit_intercept=False)
regressor.fit(x, y)
y_pred = regressor.predict(x)

plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.title('Number of images vs Monthly Sales')
plt.xlabel('Number of Images')
plt.ylabel('Monthly Sales')
plt.show()
```

```python
#Multiple Linear Regression 
x = df_norb[['Number_images','Reviews_Rating', 'Price']]
y = df_norb[['MonthlySales']]

reg = LinearRegression(fit_intercept=False)
reg.fit(x,y)
predictedsales = reg.predict(x)
```

```python
X0 = np.array(x['Number_images'])
plt.scatter(X0, y, color = 'red')
plt.scatter(X0, reg.predict(x), color = 'blue')
plt.title('Number of images vs Monthly Sales')
plt.xlabel('Number of Images')
plt.ylabel('Monthly Sales')
plt.show()
```

```python
X1 = np.array(x['Reviews_Rating'])
plt.scatter(X1, y, color = 'red')
plt.scatter(X1, reg.predict(x), color = 'blue')
plt.title('Reviews Rating vs Monthly Sales')
plt.xlabel('Reviews Rating')
plt.ylabel('Monthly Sales')
plt.show()
```

```python
X2 = np.array(x['Price'])
plt.scatter(X2, y, color = 'red')
plt.scatter(X2, reg.predict(x), color = 'blue')
plt.title('Price vs Monthly Sales')
plt.xlabel('Price')
plt.ylabel('Monthly Sales')
plt.show()
```

```python
plt.scatter(y, predictedsales, color = 'blue')
plt.title('Predicted vs Actual Monthly Sales')
plt.xlabel('Actual Monthly Sales')
plt.ylabel('Predicted Monthly Sales')
plt.xlim(0,2000)
plt.ylim(0,2000)
plt.show()
```

# Finding Keywords to use on Amazon

need to count all the specific words then take out the common words

```python
textstring = ""
for c in df_norb['Title']:
    for word in c.split(" "):
        if word.isalpha() and word != 'with' and word != 'for' and word != 'and' and word != 'nm' and word != 'by' and word != 'the' and word != 'to' and word != 'up' and len(word)>1:
            textstring = textstring + word + ' '
            
print (textstring)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width = 1000, height = 500).generate(textstring)

plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
```
