```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 0.01 #amplitude of oscillator in meters
f = 10.0 #frequency of oscillator in Hz
e = 0.55 #coefficent of restitution
g = 9.81 #acceleration due to gravity

y0 = []
ym = []
v0 = []
vm = []
time = []

nsteps = 10000 #number of time steps
toff = 0.0 #offset in time since last collision
trange = 1.0 #total recording time in seconds

h00 = 0.03 #initial position of bb
v00 = 0.0 #initial velocity of bb
```

```python
for i in range(0,nsteps+1):
    time.append(i*trange/nsteps) #time in seconds
    ym.append(A*math.sin(2.0*math.pi*f*time[i]))
    vm.append(A*2.0*math.pi*f*math.cos(2.0*math.pi*f*time[i]))
    trel = time[i]-toff
    y0.append(h00 +v00*trel - 0.5*g*trel*trel)
    v0.append(v00 - g*trel)
    if (y0[i] < ym[i]):
        vrel = v0[i]-vm[i]
        v00 = -e*vrel
        h00 = ym[i]
        toff = time[i]
```

```python
fig = plt.figure()
plt.plot(time,ym,'-',time,y0,'--')
plt.legend(['Oscillator ','BB '])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (m)')
plt.show()
```
