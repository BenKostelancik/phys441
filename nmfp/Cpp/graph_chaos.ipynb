```python
from ROOT import TCanvas, gPad
from ROOT import TH1F, TGraph, TGraphErrors
from ROOT import gStyle
from array import array
import math
import random

c1 = TCanvas( 'c1', 'Simple Chaos Example', 200, 10, 1000, 800 )
c1.SetGridx()
c1.SetGridy()
c1.GetFrame().SetFillColor( 42 )
c1.GetFrame().SetBorderMode(-1 )
c1.GetFrame().SetBorderSize( 5 )
```

```python
def growthfunction(x,par):
    return par[0]*x*(1.0-x)
```

```python
gStyle.SetOptFit(1)
```

```python
lambda_start = 3.5
npoints = 40
xstart = 0.5
```

```python
population = array('d')
year = array('d')
lmbda = array('d')
xvalue = array('d')

population.append(xstart)
lmbda.append(lambda_start)
year.append(0)

n = npoints
for i in range(1,n):
    year.append(i)
    x = population[i-1]
    population.append(growthfunction(x,lmbda))
```

```python
gr = TGraph(n,year,population);

gr.SetLineColor(2);
gr.SetLineWidth(4);
gr.SetMarkerColor(4);
gr.SetMarkerStyle(21);
gr.SetMarkerSize(0.4)
gr.SetTitle("Growth Function: lambda*x*(1-x)");
gr.GetXaxis().SetTitle("Year");
gr.GetYaxis().SetTitle("Population");


c1.cd(1);
gr.Draw("APL");
```

```python
c1.Draw()
```
