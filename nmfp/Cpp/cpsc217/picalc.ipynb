```python
from math import *

papprox = []
iterm = []
papprox.append(float(3))
iterm.append(1)

print ("Iteration 1: pi = %.12g" % papprox[0])

for i in range(1,150):
    iterm.append(i+1)
    denom = (2*i)*(2*i+1)*(2*i+2)
    term = pow(-1,i+1)*4.0/denom
    papprox.append(papprox[i-1] + term)
    print ("Iteration %d: pi = %.12g" % (i+1,papprox[i]))
```

```python
import matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(iterm[100:], papprox[100:])

ax.set(xlabel='Term Number', ylabel='Pi',
       title='Simple Series Apprximation of Pi')
ax.grid()

plt.show()
```
